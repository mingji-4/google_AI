```python
import random
from PIL import Image
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display, clear_output
```

```python
strawberry_images = [
    ("/content/red_strawberry-140_jpg.rf.830c553f17b016313b684122d51f3f9b.jpg", "잘익은 딸기"),
    ("/content/green_strawberry-030_jpg.rf.d6393078c8a83605b291fa325a44e4c9.jpg", "덜익은 딸기"),
    ("/content/black_strawberry-001_jpg.rf.821b0c0340c3cd42db88874ff5898f0e.jpg", "썩은 딸기")
]
```

```python
def play_strawberry_classification_game():
    print("딸기 분류 게임을 시작합니다!")
    print("게임에서 보여지는 딸기는 잘익은 딸기, 덜익은 딸기, 썩은 딸기 중 하나입니다.")
    print("딸기의 상태를 맞춰주세요.")

    strawberry_image_path, correct_state = random.choice(strawberry_images)

    img = Image.open(strawberry_image_path)
    plt.imshow(img)
    plt.axis('off')
    plt.show()

    input_box = widgets.Text(placeholder='정답을 입력하세요', description='답:')
    display(input_box)

    def check_answer(b):
        user_guess = input_box.value.strip()
        clear_output()
        if user_guess == correct_state:
            print("정답입니다! 이 딸기는", correct_state, "입니다.")
        else:
            print("틀렸습니다. 정답은", correct_state, "입니다.")

    submit_button = widgets.Button(description="제출")
    submit_button.on_click(check_answer)
    display(submit_button)
```

## 간단한 딸기 분류 게임

```python
play_strawberry_classification_game()
```
